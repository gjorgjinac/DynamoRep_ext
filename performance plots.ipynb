```python
budget=20
precision=get_precision(df,dimension,budget)
fig, axes=plt.subplots(nrows=6, ncols=4, sharex=True, figsize=(15,22))
for problem_id in range(1,25):
    ax=axes[int((problem_id-1)/4)][(problem_id-1)%4]
    sns.lineplot(precision.query('problem_id==@problem_id'), x='instance_id',y='precision', hue='algorithm_name', ax=ax, palette=color_palette[0:len(all_algorithms)])
    ax.set_title(f'Problem class: {problem_id}')
    if problem_id==1:
        ax.legend(title='Algorithm Name')
    else:
        ax.get_legend().remove()
plt.tight_layout()
plt.show()
```

```python
grid = sns.FacetGrid(precision, col="problem_id", hue="algorithm_name", palette="tab20c",
                     col_wrap=4, height=1.5)

grid.map(plt.plot, "instance_id", "precision")


# Adjust the arrangement of the plots
grid.fig.tight_layout(w_pad=1)
```
