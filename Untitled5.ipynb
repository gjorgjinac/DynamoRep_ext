```python
count_missing={column:all_ela_features[column].isna().sum() for column in all_ela_features.columns}
count_missing=pd.DataFrame([count_missing]).T
count_missing.columns=['missing']
count_missing['missing_percent']=count_missing['missing'].apply(lambda x: x/all_ela_features.shape[0])
id_columns=['problem_id','instance_id','algorithm_name','iteration','dim','seed']
columns_to_keep=set(count_missing.query('missing_percent<0.1').index).difference(set(id_columns))
pd.DataFrame(columns_to_keep,columns=['feature']).to_csv('ela_problem_classification_visualizations/ela_to_keep.csv')
```
